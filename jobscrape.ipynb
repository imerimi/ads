### Scrape Data and Store in DF

Scrape IT related job within 3 days.

Columns include:
1. title
2. company
3. location
4. category
5. job type
6. salarylist
7. job description

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from bs4 import BeautifulSoup
import re


# Set the user agent string
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36"

# Set Chrome options for headless mode and user agent
chrome_options = Options()
chrome_options.add_argument(f'user-agent={user_agent}')

# Start a new Chrome session with the configured options
driver = webdriver.Chrome(options=chrome_options)



jobid=[]


for i in range(10):
    # Open the JobStreet website
    driver.get(f"https://www.jobstreet.com.my/jobs-in-information-communication-technology/in-Malaysia?daterange=3&page={i+1}")
    
    soup = BeautifulSoup(driver.page_source,'html.parser')
#     print(soup)

    test = soup.find(string=re.compile("jobIds"))
    test2 = test.split('jobIds":[')[1]
    # test2
    test3 = test2.split(']')[0]
    # test3
    test3= test3.replace("\"","")
    # test3
    key3= test3.split(',')
    for i in key3:
        jobid.append(i)
        
print(len(jobid))
for i in jobid:
    print(i)
    
```

```python
titlelist=[]
companylist=[]
locationlist=[]
catlist=[]
modelist=[]
salarylist=[]
jdlist=[]
```

```python
import time

# Set the user agent string
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36"

# Set Chrome options for headless mode and user agent
chrome_options = Options()
chrome_options.add_argument(f'user-agent={user_agent}')

# Start a new Chrome session with the configured options
driver = webdriver.Chrome(options=chrome_options)
# Open the JobStreet website
for i in jobid:
    driver.get(f"https://www.jobstreet.com.my/job/{i}")

    soup = BeautifulSoup(driver.page_source,'html.parser')
    print(f"working on {i}")
    time.sleep(8)
    # print(soup)
    title2 = soup.find('h1')
    title = title2.get_text()
    titlelist.append(title)

    company2 = soup.find('span',attrs={"data-automation": "advertiser-name"})
    company = company2.get_text()
    companylist.append(company)
    
    maindata=[]
    for data in soup.find('div',attrs={"class": "y735df0 _1akoxc50 _1akoxc56"}):
        maindata.append(data.find('span').text)

    locationlist.append(maindata[0])
    catlist.append(maindata[1])
    modelist.append(maindata[2])
    if len(maindata)>3:
        salarylist.append(maindata[3])
    else:
        salarylist.append(None)


    for data in soup.find('div',attrs={"data-automation": "jobAdDetails"}):
    #     print(data.text)
        jdlist.append(data.text)
```

```python
import pandas as pd
df = pd.DataFrame({
    'Title': titlelist[0:318],
    'Company': companylist,
    'Location': locationlist,
    'Category':catlist,
    'Work Type':modelist,
    'Salary':salarylist,
    'Job Description':jdlist
})

df
```

### Data Cleaning

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
def remove_non_english(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)

df['Title'] = df['Title'].apply(remove_non_english)
df['Title'] = df['Title'].str.upper()
df['Title'] = df['Title'].str.replace(r'\(.*?\)', '', regex=True)
df
```

```python
df['Main Location']=""

df['location'] = df['location'].str.upper()
for i in range(len(df['location'])):
    if 'SELANGOR' in df.iloc[i]['location']:
        df['Main Location'][i]="SELANGOR"
    elif 'KUALA LUMPUR'in df.iloc[i]['location']:
        df['Main Location'][i]="KUALA LUMPUR"
    elif 'PENANG'in df.iloc[i]['location']:
        df['Main Location'][i]="PENANG"
    elif 'JOHOR'in df.iloc[i]['location']:
        df['Main Location'][i]="JOHOR"
    elif 'MELAKA'in df.iloc[i]['location']:
        df['Main Location'][i]="MELAKA"
    elif 'SARAWAK'in df.iloc[i]['location']:
        df['Main Location'][i]="SARAWAK"
    elif 'PUTRAJAYA'in df.iloc[i]['location']:
        df['Main Location'][i]="PUTRAJAYA"
    elif 'SABAH'in df.iloc[i]['location']:
        df['Main Location'][i]="SABAH"
    elif 'PERAK'in df.iloc[i]['location']:
        df['Main Location'][i]="PERAK"
    elif 'NEGERI SEMBILAN'in df.iloc[i]['location']:
        df['Main Location'][i]="NEGERI SEMBILAN"
    elif 'KELANTAN'in df.iloc[i]['location']:
        df['Main Location'][i]="KELANTAN"
    elif 'KEDAH'in df.iloc[i]['location']:
        df['Main Location'][i]="KEDAH"
    

df
        
```

```python
df['Category'] = df['Category'].str.upper()
df['Company'] = df['Company'].str.upper()
df['Work Type'] = df['Work Type'].str.upper()
df['Job Description'] = df['Job Description'].str.upper()
df['Job Description'] = df['Job Description'].apply(remove_non_english)

df
```

```python
df['Avg Salary']= 0
df['Salary'] = df['Salary'].str.upper()
df['Salary'] = df['Salary'].str.replace('K', '000')
start=0
end=0
part1=""

def remove_non_numeric_and_convert_to_int(s):
    # Remove English characters and spaces using regular expressions
    s = re.sub(r'[a-zA-Z\s]', '', s)
    s = s.replace(',',"")
    # Convert the resulting string to an integer
    return int(s)

for i in range(len(df['Salary'])):
    salary = df.iloc[i]['Salary']
    
    if salary is not None and "MONTH" in salary:
        salary= salary.replace(" ","")
        salarysplit= salary.split('–')
        start = remove_non_numeric_and_convert_to_int(salarysplit[0])
        end = remove_non_numeric_and_convert_to_int(salarysplit[1])
#         print(salary,"\n", start,"\n",end,"\n",(start+end)/2)
        df['Avg Salary'][i]= (start+end)/2
        df['Salary'][i]= (f"{start}-{end}")
    
    if salary is not None and "MYR" in salary:
        salary= salary.replace(" ","")
        salarysplit= salary.split('-')
        start = remove_non_numeric_and_convert_to_int(salarysplit[0])
        end = remove_non_numeric_and_convert_to_int(salarysplit[1])
#         print(salary,"\n", start,"\n",end,"\n",(start+end)/2)
        df['Avg Salary'][i]= (start+end)/2
        df['Salary'][i]= (f"{start}-{end}")
    
    if salary is not None and "YEAR" in salary :
        salary= salary.replace(" ","")
        salarysplit= salary.split('–')
        start = remove_non_numeric_and_convert_to_int(salarysplit[0])
        end = remove_non_numeric_and_convert_to_int(salarysplit[1])
#         print(salary,"\n", start,"\n",end,"\n",(start+end)/2)
        df['Avg Salary'][i]= (start/12+end/12)/2
        df['Salary'][i]= (f"{start}-{end}")
    
    if salary is not None and "P.A" in salary:
        salary= salary.replace(" ","")
        salary= salary.replace("$","")
        salarysplit= salary.split('-')
        start = remove_non_numeric_and_convert_to_int(salarysplit[0])
        end = int(salarysplit[1].replace('P.A.',""))
        df['Avg Salary'][i]= (start/12+end/12)/2
        df['Salary'][i]= (f"{start}-{end}")
        
    if salary is not None and "P.M." in salary:
        salary= salary.replace(" ","")
        salary= salary.replace(".0","")
        salary= salary.replace("$","")
        salary= salary.replace(".00","")
        salarysplit= salary.split('-')
        if len(salarysplit) !=2:
            salarysplit= salary.split('−')
        start = int(salarysplit[0])
        part1 = salarysplit[1].split('P')
        end= int(part1[0])
        df['Avg Salary'][i]= (start+end)/2
        df['Salary'][i]= (f"{start}-{end}")

df
```
